```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Blue" : [30,40,20,60,50,25,35,45,55,65],
    "Green" : [50,60,40,80,70,45,55,65,75,85],
    "Red" : [40,90,35,100,80,30,50,70,85,95],
    "NIR" : [150,80,160,90,200,140,170,110,130,120]
}

df = pd.DataFrame(data)

df
```

```python
df["NDVI"] = (df["NIR"] - df["Red"])/(df["NIR"] + df["Red"])
df
```

```python
df.describe()
```

```python
plt.hist(df["NDVI"],bins = 5)

plt.title("NDVI Discribution")
plt.xlabel("NDVI")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.scatter(df["Red"],df["NIR"])

plt.xlabel("Red")
plt.ylabel("NIR")
plt.title("Red vs NIR")

plt.show()
```

```python
print(df.corr())
```

```python
import seaborn as sns

sns.heatmap(df.corr(),
            annot = True,
            cmap = "coolwarm")
plt.title("Band Correlation")
plt.show()
```
