```python
import plotly.graph_objects as go
import numpy as np
```

```python
u = np.linspace(0,2*np.pi,100)
v = np.linspace(0,np.pi,100)

x = np.outer(np.cos(u),np.sin(v))
y = np.outer(np.sin(u),np.sin(v))
z = np.outer(np.ones(np.size(u)),np.cos(v))

fig = go.Figure(
    data = [
        go.Surface(
            x = x,
            y = y,
            z = z
        )
    ]
)

fig.update_layout(
    title = "GeoSnap Earth",
    width = 800,
    height = 800
)

fig.show()
```
