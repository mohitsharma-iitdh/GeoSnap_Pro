```python
import pandas as pd 

data = {
    "Red" : [25,110,45,30,95],
    "NIR" : [120,90,150,40,180]
}

df = pd.DataFrame(data)
df
```

```python
df["NDVI"] = (df["NIR"] - df["Red"]) / (df["NIR"] + df["Red"])
df
```

```python
for i in range(len(df)):
    if df.loc[i,"NDVI"] > 0.5:
        print(f"Pixel {i}: Healthy Vegetation")
    elif df.loc[i,"NDVI"] > 0:
        print(f"Pixel {i}: Sparse Vegetation")
    else:
        print(f"Pixel {i}: Non-Vegetation / Water / Urban")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(range(len(df)),df["NDVI"])

plt.title("NDVI Values")
plt.xlabel("Pixel Index")
plt.ylabel("NDVI")

plt.show()
```

```python
df["Class"] = df["NDVI"].apply(
    lambda x: "Healty Vegetation"
    if x > 0.5
    else "Sparse Vegetation"
    if x > 0
    else "Non-Vegetation"
)

df
```
